```python
# Factorizar
data["job_f"] = pd.factorize(data["job"])[0]
data["marital_f"] = pd.factorize(data["marital"])[0]
data["education_f"] = pd.factorize(data["education"])[0]
data['default_f']=pd.factorize(data["default"])[0]
data['housing_f']=pd.factorize(data["housing"])[0]
data['loan_f']=pd.factorize(data["loan"])[0]
data['month_f']=pd.factorize(data["month"])[0]
data['poutcome_f']=pd.factorize(data["poutcome"])[0]
data['y_f']=pd.factorize(data["y"])[0]
```

```python
fig, axis = plt.subplots(figsize = (12, 6))
matriz_corr=data[['job_f',"age",'marital_f','y_f','education_f','default_f','housing_f','loan_f','month_f',"duration",'poutcome_f', "campaign"]].corr()
#print(matriz_corr)
sns.heatmap(matriz_corr, annot = True, fmt = ".2f")

plt.tight_layout()

plt.show()
```
