```python
# Analisis de Variables categoricas relacionadas a la data

fig, axis = plt.subplots(5, 2, figsize = (20, 15))

# Crear un histograma mÃºltiple
sns.histplot(ax = axis[0, 0], data = data, x = "education")
sns.histplot(ax = axis[0, 1], data = data, x = "education", stat="percent")
sns.histplot(ax = axis[1, 0], data = data, x = "job")
sns.histplot(ax = axis[1, 1], data = data, x = "job", stat="percent")
sns.histplot(ax = axis[2, 0], data = data, x = "marital").set_xticks([])
sns.histplot(ax = axis[2, 1], data = data, x = "marital", stat="percent").set_xticks([])
sns.histplot(ax = axis[3, 0], data = data, x = "loan")
sns.histplot(ax = axis[3, 1], data = data, x = "loan", stat="percent")
sns.histplot(ax = axis[4, 0], data = data, x = "default")
sns.histplot(ax = axis[4, 1], data = data, x = "default", stat="percent")



# Ajustar el layout
plt.tight_layout()

# Mostrar el plot
plt.show()
```

```python
# Analisis de variables numericas

fig, axis = plt.subplots(4, 2, figsize = (10, 14), gridspec_kw = {"height_ratios": [6, 1, 6, 1]})

# Crear una figura mÃºltiple con histogramas y diagramas de caja
sns.histplot(ax = axis[0, 1], data = data, x = "age")
sns.boxplot(ax = axis[1, 0], data = data, x = "age")
sns.histplot(ax = axis[0, 1], data = data, x = "duration")
sns.boxplot(ax = axis[1, 1], data = data, x = "duration")
sns.histplot(ax = axis[2, 0], data = data, x = "campaign")
sns.boxplot(ax = axis[3, 0], data = data, x = "campaign")


# Ajustar el layout
plt.tight_layout()
fig.delaxes(axis[2, 1])
fig.delaxes(axis[3, 1])

# Mostrar el plot
plt.show()
```
